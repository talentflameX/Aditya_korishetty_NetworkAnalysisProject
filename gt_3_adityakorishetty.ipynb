In the intended network graph of "Adventures of Huckleberry Finn," each node represents a character, and each edge represents an interaction between characters. The size and color of each node were to be determined by the degree centrality, reflecting the character's importance and frequency of interactions within the story.

Key Points from the Analysis:

Central Characters: Characters like Huck Finn, Jim, and Tom Sawyer were expected to have larger nodes due to their high centrality, indicating their pivotal roles in the novel's social network.
Peripheral Characters: Characters with fewer interactions, such as the Shepherdsons or the Wilks sisters, would have smaller nodes, indicating their less central role.
Relationship Dynamics: The graph would visually depict the complex relationships among characters, like alliances and conflicts.
Community Clusters: Any clusters or communities detected would reflect groups or alliances within the story, such as the feud between the Grangerfords and Shepherdsons.

```python
# 1b,c,d
#!pip install networkx
#!pip install matplotlib
#!pip install community
#!pip install python-louvain
!pip install scipy
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

#Creating a new graph
G = nx.Graph()

#Adding nodes (characters)
characters = ["Huck Finn", "Jim", "Tom Sawyer", "Pap Finn", "Widow Douglas", "Miss Watson", 
              "The Duke", "The King", "Judge Thatcher", "Aunt Polly", "Sid Sawyer", "Mary Jane Wilks", 
              "Susan Wilks", "Joanna Wilks", "Peter Wilks", "Sally Phelps", "Silas Phelps", 
              "Ben Rogers", "Joe Harper", "Colonel Grangerford", "Buck Grangerford", "Shepherdsons"]

G.add_nodes_from(characters)

#Adding edges (interactions)
interactions = [("Huck Finn", "Jim"), ("Huck Finn", "Tom Sawyer"), ("Huck Finn", "Pap Finn"),
                ("Huck Finn", "Widow Douglas"), ("Huck Finn", "Miss Watson"), ("Jim", "Tom Sawyer"),
                ("The Duke", "The King"), ("Huck Finn", "The Duke"), ("Huck Finn", "The King"),
                ("Tom Sawyer", "Aunt Polly"), ("Tom Sawyer", "Sid Sawyer"), ("Mary Jane Wilks", "Susan Wilks"),
                ("Susan Wilks", "Joanna Wilks"), ("Huck Finn", "Mary Jane Wilks"), ("Huck Finn", "Susan Wilks"),
                ("Huck Finn", "Sally Phelps"), ("Sally Phelps", "Silas Phelps"), ("Huck Finn", "Ben Rogers"),
                ("Huck Finn", "Joe Harper"), ("Colonel Grangerford", "Buck Grangerford"),
                ("Buck Grangerford", "Shepherdsons"), ("Huck Finn", "Colonel Grangerford")]

G.add_edges_from(interactions)

#Calculating degree centrality
centrality = nx.degree_centrality(G)

#Assigning node size based on degree centrality (importance)
node_size = [v * 3000 for v in centrality.values()]

#Assigning colors based on a simple rule (for illustration)
colors = ["blue" if node in ["Huck Finn", "Jim", "Tom Sawyer"] else "green" for node in G.nodes()]

#Drawing the network
plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_color=colors, node_size=node_size, edge_color="gray", font_size=8)
plt.title("Network Graph of 'Adventures of Huckleberry Finn'")
plt.show()



#Assuming G is your graph

#Calculate Degree Centrality
degree_centrality = nx.degree_centrality(G)

#Calculate Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)

#Calculate Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)

#Calculate PageRank
pagerank = nx.pagerank(G)

#Print the centralities for main protagonists
main_protagonists = ["Huck Finn", "Jim", "Tom Sawyer", "Pap Finn"]
for protagonist in main_protagonists:
    print(f"Centrality measures for {protagonist}:")
    print(f"Degree Centrality: {degree_centrality.get(protagonist, 'Not found')}")
    print(f"Betweenness Centrality: {betweenness_centrality.get(protagonist, 'Not found')}")
    print(f"Closeness Centrality: {closeness_centrality.get(protagonist, 'Not found')}")
    print(f"PageRank: {pagerank.get(protagonist, 'Not found')}\n")
main_protagonists = ["Huck Finn", "Jim", "Tom Sawyer", "Pap Finn"]

for protagonist in main_protagonists:
    # Extract the ego network
    ego_net = nx.ego_graph(G, protagonist)

    # Calculate the local clustering coefficient
    clustering_coefficient = nx.clustering(ego_net, protagonist)

    print(f"Local clustering coefficient for {protagonist}: {clustering_coefficient}")
```

```python
#2b
1. Fictional vs. Factual Basis
Fictional Nature: The graph is based on a work of fiction by Mark Twain. Relationships and interactions between characters are crafted by the author and may not always reflect real-world social dynamics.
Artistic License: In fiction, character interactions can be exaggerated or understated for narrative purposes, unlike real-world social networks where interactions are not predetermined.
2. Properties of Real-World Graphs
Several properties are typical of real-world social networks. Let's see if the social graph from "Adventures of Huckleberry Finn" satisfies these:

Small-World Property: Real-world social networks often exhibit small-world properties, characterized by short path lengths between nodes and a high degree of clustering. The novel's social graph might show small-world characteristics if characters (nodes) are closely connected with few intermediaries (short path lengths) and if there are tight-knit groups (high clustering).
Scale-Free Property: Many real-world networks follow a power-law distribution in node connectivity, meaning a few nodes (hubs) have many connections, while most nodes have few. The novel's graph may exhibit this if central characters like Huck Finn and Tom Sawyer have significantly more connections than other, more peripheral characters.
Community Structure: Real-world networks often show a strong community structure, where nodes are more densely connected within certain groups than between groups. The novel's social graph might display this if characters form distinct clusters or communities based on their interactions.
Preferential Attachment: In some networks, new nodes prefer to connect to already well-connected nodes. This might be less applicable to a fictional narrative, where relationships are determined by the story rather than emerging organically.
```
